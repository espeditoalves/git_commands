# Análise de Dados Simples
Este notebook carrega um dataset CSV, mostra um resumo básico e salva resultados em `scripts/output`.

```python
from pathlib import Path
from dataset.data_utils import load_dataset, summarize_dataset
from scripts import describe_module
from scripts.input.data_loader import load_csv
from scripts.output.data_exporter import export_summary_csv, save_histogram

print(describe_module())
```

```python
dataset_path = Path('dataset/supplement_impact_data.csv')
try:
    df = load_dataset(dataset_path)
except FileNotFoundError:
    print(f'Dataset não encontrado em {dataset_path}. Coloque o CSV na pasta dataset/ e reinicie.')
    df = None

if df is not None:
    print('Primeiras linhas do dataset:')
    display(df.head())
    summary = summarize_dataset(df)
    print('Resumo de colunas e valores ausentes:')
    print(summary['missing_values'])
```

```python
if df is not None:
    output_csv = Path('scripts/output/summary.csv')
    export_summary_csv(df, output_csv)
    print(f'Resumo exportado para: {output_csv}')

    image_path = Path('image/data_histogram.png')
    save_histogram(df, df.columns[0], image_path)
    print(f'Gráfico salvo em: {image_path}')
```
